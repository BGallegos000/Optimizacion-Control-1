#Cual es el mayor de la lista {1, 3, 5, 9, 11, 3, 2}

```python
lista_numeros = [1, 3, 5, 9, 11, 3, 2]

mayor_actual = lista_numeros[0]

for numero in lista_numeros:
    if numero > mayor_actual:
        mayor_actual = numero

print(f"El número mayor es: {mayor_actual}")
```

# Versión ampl

```python
model_mayor = """
set INDICES;
param numero{INDICES};

var x{INDICES} binary;

maximize Numero_Mayor:
    sum{i in INDICES} numero[i] * x[i];

subject to Seleccionar_Solo_Uno:
    sum{i in INDICES} x[i] = 1;

# Bloque de datos integrado
data;
set INDICES := 1 2 3 4 5 6 7;
param numero :=
1 1
2 3
3 5
4 9
5 11
6 3
7 2;
"""
```

```python
ampl.reset()
ampl.eval(model_mayor)
ampl.option["solver"] = "highs"
ampl.solve()

print("El mayor número es: ", ampl.get_objective('Numero_Mayor').value())
```

# Cual es el que más se repite {a, a, b, b, #, #, 3, 3, 3, l, k, L, m}

```python
lista_caracteres = ['a', 'a', 'b', 'b', '#', '#', '3', '3', '3', 'l', 'k', 'L', 'm']

frecuencias = {}

for char in lista_caracteres:
    if char in frecuencias:
        frecuencias[char] += 1
    else:
        frecuencias[char] = 1

ganador = None
max_votos = 0

for char, votos in frecuencias.items():
    if votos > max_votos:
        max_votos = votos
        ganador = char

print(f"El carácter más repetido es: '{ganador}' (Aparece {max_votos} veces)")
```

# Versión ampl

```python
model_cuantoserepite = """
set CARACTERES;
param frecuencia{CARACTERES};

var x{CARACTERES} binary;

maximize Mas_Repetido:
    sum{c in CARACTERES} frecuencia[c] * x[c];

subject to Seleccionar_Solo_Uno:
    sum{c in CARACTERES} x[c] = 1;

# Bloque de datos integrado
data;
set CARACTERES := a b hash 3 l k L m;
param frecuencia :=
a    2
b    2
hash 2
3    3
l    1
k    1
L    1
m    1;
"""
```

```python
ampl.reset()
ampl.eval(model_cuantoserepite)
ampl.option["solver"] = "highs"
ampl.solve()

print("El caracter que más se repite es: ", ampl.get_objective('Mas_Repetido').value())
```
