## Ejercicio N°5 

* La empresa Transporte y Tránsito Global (TTG) estudia la factibilidad de
introducir un sistema de autobuses de transporte masivo, que aliviará el
tránsito de la ciudad. El estudio inicial busca determinar el mínimo número
de autobuses que pueden suplir las necesidades de transporte en la ciudad.
Luego de un análisis de datos, se identificó que el número de autobuses
requerido se puede suponer constante para intervalos de cuatro horas, como
se presenta en la Tabla La empresa TTG decidió que, para hacer el
mantenimiento diario requerido, cada autobús debe operar solo ocho horas
consecutivas al día. ¿Cuántos buses se deben asignar y en qué turnos deben
trabajar para cumplir con la demanda al mínimo costo?




```python
model5 = """

    var x1 >= 0 integer;
var x2 >= 0 integer;
var x3 >= 0 integer;
var x4 >= 0 integer;
var x5 >= 0 integer;
var x6 >= 0 integer;

minimize z:
x1 + x2 + x3 + x4 + x5 + x6;

s.t.
r1: x1 + x6 >= 14;
r2: x1 + x2 >= 6;
r3: x2 + x3 >= 12;
r4: x3 + x4 >= 5;
r5: x4 + x5 >= 11;
r6: x5 + x6 >= 8;

"""

ampl.reset()
ampl.eval(model5)
ampl.option["solver"] = 'highs'
ampl.solve()
print(f"Turno 12am - 4pm: {ampl.get_value('x1')}")
print(f"Turno 4am - 8pm: {ampl.get_value('x2')}")
print(f"Turno 8am - 12pm: {ampl.get_value('x3')}")
print(f"Turno 12pm - 4pm: {ampl.get_value('x4')}")
print(f"Turno 4pm - 8pm: {ampl.get_value('x5')}")
print(f"Turno 8mm - 12am: {ampl.get_value('x6')}")
print(f"Mejor asignacion de turnos minimizando: {ampl.get_value('z')}")
```
